## 기존 평가완료 데이터를 **태블로용** 키워드 데이터로 만드는 코드

```python
import pandas as pd
import os
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
# 상위 디렉토리의 경로를 만듦
parent_directory = os.path.dirname(os.getcwd())

# CSV 파일 불러오기
evaluation = pd.read_csv(os.path.join(parent_directory, 'data', '어린이집_total - 어린이집_평가완료.csv'), encoding='UTF-8')
evaluation.head()
```

```python
data_dic = {'어린이집코드' : [], '어린이집' : [], '어린이집 주소' : [], '카테고리' : [], '키워드' : [], '긍부정' : []}
```

```python
evaluation_list = ['보육 과정 및 상호작용', '보육환경 및 운영관리', '건강 ･안전', '교직원']
for x in range(len(evaluation)):
    temp = evaluation.loc[x, ['어린이집코드', '어린이집', '어린이집 주소', '보육 과정 및 상호작용 긍정키워드', '보육 과정 및 상호작용 부정키워드', '보육환경 및 운영관리 긍정키워드', '보육환경 및 운영관리 부정키워드', '건강 ･안전 긍정키워드', '건강 ･안전 부정키워드', '교직원 긍정키워드', '교직원 부정키워드']].copy()
    for element in evaluation_list:
        if not pd.isna(temp[element + ' 긍정키워드']):
            p_word = temp[element + ' 긍정키워드'].split(' ')
            for p in p_word:
                data_dic['어린이집코드'].append(temp['어린이집코드'])
                data_dic['어린이집'].append(temp['어린이집'])
                data_dic['어린이집 주소'].append(temp['어린이집 주소'])
                data_dic['카테고리'].append(element)
                data_dic['키워드'].append(p)
                data_dic['긍부정'].append('긍정')
                
        if not pd.isna(temp[element + ' 부정키워드']):
            n_word = temp[element + ' 부정키워드'].split(' ')
            for n in n_word:
                data_dic['어린이집코드'].append(temp['어린이집코드'])
                data_dic['어린이집'].append(temp['어린이집'])
                data_dic['어린이집 주소'].append(temp['어린이집 주소'])
                data_dic['카테고리'].append(element)
                data_dic['키워드'].append(n)
                data_dic['긍부정'].append('부정')
                
            
            
```

```python
df = pd.DataFrame(data_dic)
df
```
