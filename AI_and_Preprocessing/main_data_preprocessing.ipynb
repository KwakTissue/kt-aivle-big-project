# 어린이집 total 데이터 형성

```python
import pandas as pd
import re
from datetime import datetime
import os
```

```python
pd.options.display.float_format = '{:.5f}'.format
```

```python
# 상위 디렉토리의 경로를 만듦
parent_directory = os.path.dirname(os.getcwd())

# CSV 파일 불러오기
nursery_seoul = pd.read_csv(os.path.join(parent_directory, 'data', '서울시 어린이집 정보(표준 데이터).csv'), encoding='cp949')
nursery_score = pd.read_csv(os.path.join(parent_directory, 'data', '어린이집평가 데이터_주소포함.csv'), encoding='UTF8')
```

# 어린이집 정보 데이터

```python
nursery_seoul.columns
```

```python
nursery_seoul_total = nursery_seoul[['시군구명', '어린이집코드', '어린이집명', '어린이집유형', '운영현황', '우편번호',
                                      '상세주소', '전화번호', '시설 위도(좌표값)', '시설 경도(좌표값)', '제공서비스']]
```

```python
# 폐업된 어린이집 제외
nursery_seoul_total = nursery_seoul_total[nursery_seoul_total['운영현황']=='정상'].reset_index(drop=True)
nursery_seoul_total
```

```python
nursery_seoul_total.info()
```

```python
# 우편번호 결측치 행 확인
nursery_seoul_total[nursery_seoul_total['우편번호'].isnull()]
```

```python
# 우편번호 결측치 행 제거
nursery_seoul_total = nursery_seoul_total.dropna(subset='우편번호', axis=0)
nursery_seoul_total[nursery_seoul_total['우편번호'].isnull()]
```

```python
# 우편번호 타입 변경
nursery_seoul_total['우편번호'] = nursery_seoul_total['우편번호'].astype(int)
```

```python
nursery_seoul_total['어린이집코드'].nunique()
```

# 어린이집 평가 데이터 전처리

```python
nursery_score = nursery_score.drop(columns='Unnamed: 0', axis=1)
```

```python
nursery_score['어린이집명'].nunique()
```

```python
# 주소에서 우편번호 분리
nursery_score['우편번호'] = nursery_score['주소'].apply(lambda x: re.findall(r'\d+', x)[0])
# 타입 변경
nursery_score['우편번호'] = nursery_score['우편번호'].astype(int)
```

```python
nursery_score.info()
```

# 어린이집 기본정보, 평가 merge

```python
total = pd.merge(nursery_score, nursery_seoul_total, how='inner')
len(total)
```

```python
# 어린이집코드를 기준으로 중복행 확인
total[total.duplicated(subset='어린이집코드', keep=False)]
```

```python
total.head()
```

```python
# 불필요한 열 제거
total = total.drop(columns=['구', '주소', '운영현황'], axis=1)
total.head()
```

```python
# 타입 확인
total.info()
```

```python
# 평가가 없는 데이터 제거
total = total.dropna(subset=['어린이집등급', '보육과정 및 상호작용', '보육환경 및 운영관리', '건강 및 안전', '교직원', ], axis=0)
total[total['교직원'].isnull()]
```

```python
# 우편번호 결측치 행 제거
total = total.dropna(subset='평가 인증 년월', axis=0)
total[total['평가 인증 년월'].isnull()]
```

```python
# 평가 인증 년월 datetime으로 타입 변경
total['평가 인증 년월'] = total['평가 인증 년월'].astype(str)
total['평가 인증 년월'] = total['평가 인증 년월'].apply(lambda x : x.replace('.', ''))
total['평가 인증 년월'] = total['평가 인증 년월'].apply(lambda _ : datetime.strptime(_,'%Y%m'))
```

```python
total.info() # 중복행, 결측치 행 제거, 타입변경 결과 확인
```
